```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=2,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split as split
```

```python
x_train,x_test,y_train,y_test=split(x,y,test_size=0.2,random_state=0)
```

```python
model=RandomForestClassifier(n_estimators=100, min_samples_leaf=1, bootstrap=True, oob_score=True, n_jobs=-1, random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,model.predict(x_test))
cm
```

```python
model.oob_score_
```

```python
# Visualising the Train set results
plt.figure(figsize=(12,8))
from matplotlib.colors import ListedColormap
X_set, y_set = x_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
# plt.contourf(X1, X2, finalModel.predict_proba(
#     np.array([X1.ravel(), X2.ravel()]).T)[:,-1]
#     .reshape(X1.shape),
#              alpha = 0.75, cmap='jet',vmin=0,vmax=1)
# plt.colorbar()
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.1, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest Classificaiton (Train set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
